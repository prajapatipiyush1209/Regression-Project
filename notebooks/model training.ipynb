```python
#Seperate diffrentiate independent column(input) and dependent column(output)
X = data.drop(labels=['price'], axis = 1)
Y = data[['price']]
```

```python
# Diffrentiate the categorical and numerical column from the X
categorical_col = X.select_dtypes(include='object').columns
numerical_col = X.select_dtypes(exclude='object').columns
```

```python
categorical_col
```

```python
numerical_col
```

```python
#Ranking of the categorical data for the Encoding
cut_categories = ['Fair', 'Good', 'Very Good','Premium','Ideal']
color_categories = ['D', 'E', 'F', 'G', 'H', 'I', 'J']
clarity_categories = ['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']
```

```python
from sklearn.impute import SimpleImputer#For the automate the handling the missing value of the data 
from sklearn.preprocessing import StandardScaler #for the Feature Scaling 
from sklearn.preprocessing import OrdinalEncoder # For the performing the encoding of the ordinal categorical data 

from sklearn.pipeline import Pipeline
#The purpose of the pipeline is to assemble several steps that can be cross-validated together while setting different parameters. 
from sklearn.compose import ColumnTransformer
```

```python
#Now create the pipline 

num_pipeline = Pipeline(steps=
                   [('imputer',SimpleImputer(strategy='median')),
                    ('scaler', StandardScaler())])

cat_pipeline = Pipeline(
    steps = [
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ordinalencoding', OrdinalEncoder(categories=[cut_categories,color_categories,clarity_categories])),
    ('scaler', StandardScaler())

    ])

preprocessor = ColumnTransformer([
    ('num_pipeline', num_pipeline,numerical_col),
    ('cat_pipeline', cat_pipeline,categorical_col)]
)
```

```python
#Train test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=42)
```

```python
X_train.head()
```

```python
X_train=preprocessor.fit_transform(X_train)
```

```python
models = {
   'LinearRegression': LinearRegression(),
   'Lasso': Lasso(),
    'Ridge' : Ridge(),
    'Elasticnet': ElasticNet()
}
```

```python
import numpy as np
def evaluate_model(true, predicted):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2_square = r2_score(true, predicted)
    return mae, rmse, r2_square
```

```python

model_list=[]
r2_list=[]
for i in range(len(list(models.values()))):
    model = list(models.values())[i]
    model.fit(X_train,y_train)
    y_predicted = model.predict(X_test)
    
    mae, rmse, r2_square=evaluate_model(y_test,y_predicted)

    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])

    print('Model Training Performance')
    print("RMSE:",rmse)
    print("MAE:",mae)
    print("R2 score",r2_square*100)

    r2_list.append(r2_square)
    
    print('='*35)
    print('\n')

```
